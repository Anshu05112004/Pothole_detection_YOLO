```python
while True:
    s,img=cap.read()
    
    results=model(img,stream=True)
    for r in results:
        boxes=r.boxes
        for box in boxes:
                # x1,y1,x2,y2=box.xyxy[0]
                # x1,y1,x2,y2=int(x1),int(y1),int(x2),int(y2)
                # print(x1,x2,y1,y2)
                # cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,255),3)
                x1,y1,x2,y2=box.xyxy[0]
                x1,y1,w,h=int(x1),int(y1),int(x2-x1),int(y2-y1)
                
                # Confidence converted upto two decimal places
                conf=math.ceil((box.conf[0]*100))/100
                if(conf>0.45):
                    cvzone.cornerRect(img,(x1,y1,w,h))
                    cls=int(box.cls[0])
                    cv2.putText(img,f'{classNames[0]} {conf}',(x1,y1-10),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,0,255),2)    
                # Class Name
                
                
                #rectangle
                
                
                print(conf)               
    cv2.imshow('Image',img)
    cv2.waitKey(1)          
```
